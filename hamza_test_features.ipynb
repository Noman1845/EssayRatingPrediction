```python
#Takes 11 min to run...

import spacy
import pandas as pd
import swifter

nlp = spacy.load("en_core_web_sm", disable=['ner', 'parser'])

def get_doc(text):
    return nlp(text)

def verbs(doc):
    return sum(1 for token in doc if token.pos_ == "VERB")

def nouns(doc):
    return sum(1 for token in doc if token.pos_ == "NOUN")

def adverbs(doc):
    return sum(1 for token in doc if token.pos_ == "ADV")

def adjectives(doc):
    return sum(1 for token in doc if token.pos_ == "ADJ")

def pronouns(doc):
    return sum(1 for token in doc if token.pos_ == "PRON")

def interjection(doc):
    return sum(1 for token in doc if token.pos_ == "INTJ")

def propnouns(doc):
    return sum(1 for token in doc if token.pos_ == "PROPN")

# Apply the functions to the DataFrame using swifter
simplified_df['doc'] = simplified_df['essay'].swifter.apply(get_doc)
simplified_df['number_of_adjectives'] = simplified_df['doc'].swifter.apply(adjectives)
simplified_df['number_of_verbs'] = simplified_df['doc'].swifter.apply(verbs)
simplified_df['number_of_adverbs'] = simplified_df['doc'].swifter.apply(adverbs)
simplified_df['number_of_interjection'] = simplified_df['doc'].swifter.apply(interjection)
simplified_df['number_of_nouns'] = simplified_df['doc'].swifter.apply(nouns)
simplified_df['number_of_pronouns'] = simplified_df['doc'].swifter.apply(pronouns)
simplified_df['number_of_proper_nouns'] = simplified_df['doc'].swifter.apply(propnouns)

# Drop the 'doc' column if you no longer need it
simplified_df.drop(columns=['doc'], inplace=True)
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize

def words(text):

    words = word_tokenize(text)
    num_words = len(words)
    
    return num_words
   
simplified_df['number_of_words'] = simplified_df['essay'].swifter.apply(words) 



simplified_df['verbs_ratio'] = simplified_df['number_of_verbs'] / simplified_df['number_of_words']
simplified_df['adverbs_ratio'] = simplified_df['number_of_adverbs'] / simplified_df['number_of_words']
simplified_df['adjectives_ratio'] = simplified_df['number_of_adjectives'] / simplified_df['number_of_words']
simplified_df['nouns_ratio'] = simplified_df['number_of_nouns'] / simplified_df['number_of_words']
simplified_df['pronouns_ratio'] = simplified_df['number_of_pronouns'] / simplified_df['number_of_words']
simplified_df['proper_nouns_ratio'] = simplified_df['number_of_proper_nouns'] / simplified_df['number_of_words']
simplified_df['interjection_ratio'] = simplified_df['number_of_interjection'] / simplified_df['number_of_words']
```

```python
import re

# Remove HTML tags from the transition_list
transition_list = re.sub('<[^<]+?>', '', transition_list)

# Split the cleaned transition_list into a list of transition words
transitions = transition_list.split('\n')

def count_transitions(text):
    # Count the number of transition words in the text
    count = sum(1 for word in transitions if word.lower() in text.lower())
    return count

# Apply the function to the DataFrame using swifter
simplified_df['number_of_transitions_words'] = simplified_df['essay'].swifter.apply(count_transitions)
```
