```python
import pandas as pd
from features import *
import dask.dataframe as dd
import time
```

```python
df_train = pd.read_csv("../data/training_set_rel3.tsv", sep="\t", encoding="ISO-8859-1")
df_train = df_train.dropna(axis=1)
df_test = df_train#.iloc[:20]
df_test = df_test.drop(columns=["rater1_domain1", "rater2_domain1"])
df = df_test
```

```python
start = time.time()
print(f"start pipeline")

ddf = dd.from_pandas(df, npartitions=32)

df["count_characters"] = ddf["essay"].apply(lambda x: count_characters(x), meta="object")
df["count_syllables"] = ddf["essay"].apply(lambda x: count_syllables(x), meta="object")
df["count_words"] = ddf["essay"].apply(lambda x: count_words(x), meta="object")
df["count_sentences"] = ddf["essay"].apply(lambda x: count_sentences(x), meta="object")
df["flesch_reading_ease"] = ddf["essay"].apply(lambda x: get_flesch_reading_ease(x), meta="object")
df["gunning_fog"] = ddf["essay"].apply(lambda x: get_gunning_fog(x), meta="object")
df["automated_readability_index"] = ddf["essay"].apply(lambda x: get_automated_readability_index(x), meta="object")
df["smog_index"] = ddf["essay"].apply(lambda x: get_smog_index(x), meta="object")
df["flesch_kincaid_grade"] = ddf["essay"].apply(lambda x: get_flesch_kincaid_grade(x), meta="object")
df["coleman_liau_index"] = ddf["essay"].apply(lambda x: get_coleman_liau_index(x), meta="object")
df["dale_chall_readability_score"] = ddf["essay"].apply(lambda x: get_dale_chall_readability_score(x), meta="object")
df["automated_readability_index"] = ddf["essay"].apply(lambda x: get_automated_readability_index(x), meta="object")
df["dale_chall_readability_score"] = ddf["essay"].apply(lambda x: get_dale_chall_readability_score(x), meta="object")
df["difficult_words"] = ddf["essay"].apply(lambda x: get_difficult_words(x), meta="object")
df["linsear_write_formula"] = ddf["essay"].apply(lambda x: get_linsear_write_formula(x), meta="object")
df["count_awl_words"] = ddf["essay"].apply(lambda x: count_awl_words(x), meta="object")
df["calculate_lexical_diversity"] = ddf["essay"].apply(lambda x: calculate_lexical_diversity(x), meta="object")
df["get_sentence_tree_roots"] = ddf["essay"].apply(lambda x: get_sentence_tree_roots(x), meta="object")
df["get_sentence_tree_length"] = ddf["essay"].apply(lambda x: get_sentence_tree_length(x), meta="object")
df["get_average_connections_at_root"] = ddf["essay"].apply(lambda x: get_average_connections_at_root(x), meta="object")
df["get_length_of_clauses"] = ddf["essay"].apply(lambda x: get_length_of_clauses(x), meta="object")
df["calculate_misspelling_score"] = ddf["essay"].apply(lambda x: calculate_misspelling_score(x), meta="object")
df["detect_slur_usage"] = ddf["essay"].apply(lambda x: detect_slur_usage(x), meta="object")
df["calculate_overusage_of_punctuation"] = ddf["essay"].apply(lambda x: calculate_overusage_of_punctuation(x), meta="object")


tmp_df = pd.DataFrame()
tmp_df = df["essay"].apply(lambda x: get_pos_tags(x))
tmp_df = pd.json_normalize(tmp_df)
df = df.join(tmp_df, how="left")

# df["get_pos_tags"] = ddf["essay"].apply(lambda x: get_pos_tags(x), meta="object")

end = time.time()
print(f"pipeline finished in {end - start} sec.")
```
